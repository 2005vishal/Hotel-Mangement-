```python
import datetime 
import pandas as pd
import numpy as np
import random as rp
import string

booking_detail = pd.DataFrame(columns = ["Booking ID",'Name','Age','Check In','ID','Book Date','Room NO','Room Type','Transction ID','Status','Mobile','City','Amount'])
food_df = pd.DataFrame(columns = ['Transction ID','Customer ID','Date','Total_Amount']) # this dataFrame is store the customer id for food order
df_payment = pd.DataFrame(columns = ['Transaction ID','Date&Time','Customer ID','Amount','Mode','Purpose']) # this dataframe is pyment detail
non_ac     = np.arange(401,411,dtype = int)
ac         = np.arange(411,421,dtype  = int)
non_ac_3   = np.arange(421,431,dtype = int)
ac_3       = np.arange(431,441,dtype = int)
room_no_of = {'Standard Non-AC': 10,'Standard AC': 10,'3-Bed Non-AC': 10,'3-Bed AC': 10}
date_acc_room= {} # this is store availble rooms
room_num_insert = {'Standard Non-AC':non_ac.copy(),'Standard AC':ac.copy(),'3-Bed Non-AC':non_ac_3.copy(),'3-Bed AC':ac_3.copy()}
date_roomnumber = {} # this is store room number array

#this funcation is auto genarate booking 
def transction_id(cust_id,amou,mod,pur):
    global df_payment
    tr = []
    datetime_list = []
    Booking = []
    Amount =[]
    mode = []
    p =[]
    dtime = datetime.datetime.now().strftime('%d/%m/%Y %H:%M:%S')
    date = datetime.datetime.now().strftime('%Y')
    num = str(rp.randint(10000000,99999999))[:8]
    num2 = str(rp.randint(1000,9999))[:3]
    characters = string.ascii_uppercase
    char_id = ''.join(rp.choices(characters, k= 6))
    number = f'{num2}{char_id}{num}{date}'
    tr.append(number)
    datetime_list.append(dtime)
    Amount.append(amou)
    mode.append(mod)
    Booking.append(cust_id)
    p.append(pur)
    py_df = pd.DataFrame ({'Transaction ID': tr
                            ,'Date&Time':datetime_list
                            ,'Customer ID': Booking,
                            'Amount':Amount,
                            'Mode':mode
                            ,'Purpose':p})
    df_payment = pd.concat([df_payment,py_df],ignore_index = True)
    print('Payment Successful')
    return number
    
    
    
    
# transction_id('65465sdfa',2500,'Credit Card')    

def payment(cust_id,amou,pur):
    print("""
1.Credit Card
2.Debit Card
3.Net Banking
4.Cash
""")
    pur = pur
    md = int(input('Enter any option: '))
    if md == 1:
        mode = 'Credit Card'
        t_id= transction_id(cust_id,amou,mode,pur)
        return t_id
    elif md == 2:
        mode = 'Debit Card'
        t_id = transction_id(cust_id,amou,mode,pur)
        return t_id
    elif md == 3:
        mode = 'Net Banking'
        t_id=transction_id(cust_id,amou,mode,pur)
        return t_id
    elif md == 4:
        mode = 'Cash'
        t_id = transction_id(cust_id,amou,mode,pur)
        return t_id
# payment('65465sdfa',2500,'hgjh')    
    

# this funcation is given the room number according to date
def room_num_date(date,end,room_type_l):
    global room_num_insert,date_roomnumber
    date_range = pd.date_range(start=date, end=end)
    formatted_dates = [date.strftime('%d/%m/%Y') for date in date_range]
    if date in date_roomnumber.keys():
        for date in formatted_dates: 
            if date not in date_roomnumber:
                date_roomnumber[date] = room_num_insert.copy()
            else: pass
        return date_roomnumber[date][room_type_l]
    else: 
        for date in formatted_dates: 
            if date not in date_roomnumber:
                date_roomnumber[date] = room_num_insert.copy()
            else: pass
        return date_roomnumber[date][room_type_l]
# room_num_date('15/02/2024','18/02/2024','Standard Non-AC') #array([401, 402, 403, 404, 405, 406, 407, 408, 409, 410])

# this function is add room according to date  no of Room
def room_check(date,end,room_type_l):
    global room_no_of,date_acc_room
    date_range = pd.date_range(start=date, end=end)
    formatted_dates = [date.strftime('%d/%m/%Y') for date in date_range]
    if date in date_acc_room.keys():
        for date in formatted_dates: 
            if date not in date_acc_room:
                date_acc_room[date] = room_no_of.copy()
            else: pass
        return date_acc_room[date][room_type_l]
    else: 
        for date in formatted_dates: 
            if date not in date_acc_room:
                date_acc_room[date] = room_no_of.copy()
            else: pass
        return date_acc_room[date][room_type_l]
# room_check('15/02/2024','18/02/2024','Standard Non-AC') #10

#this funcation is auto genarate booking 
def book_id():
    date = datetime.datetime.now().strftime('%Y')
    num = rp.randint(10000,99999)
    num2 = rp.randint(100,999)
    characters = string.ascii_uppercase
    char_id = ''.join(rp.choices(characters, k= 6))
    return  f'{date}{char_id}{num2}'
    
    
    

def room_price(room_type_l, check_date, check_out, date_format='%d/%m/%Y'):
    date1 = datetime.datetime.strptime(check_date, date_format)
    date2 = datetime.datetime.strptime(check_out, date_format)  # Corrected typo here
    date_diff = date2 - date1
    day = date_diff.days
    if room_type_l == 'Standard Non-AC':
        return 1000 * day
    elif room_type_l == 'Standard AC':
        return 1500 * day
    elif room_type_l == '3-Bed Non-AC':
        return 1500 * day
    elif room_type_l == '3-Bed AC':
        return 2000 * day


# this code is working subtract room number in databse according to room
def room_no_subtract(date,end,room_type_l):
    global date_acc_room
    date_range = pd.date_range(start=date, end=end)
    formatted_dates = [date.strftime('%d/%m/%Y') for date in date_range]
    for i in formatted_dates:
        date_acc_room[i][room_type_l] -=1
    return 
# room_no_subtract('15/02/2024','18/02/2024','Standard Non-AC')

def room_no_to_zero(date,end,type_room,no):
    global date_roomnumber
    date_range = pd.date_range(start=date, end=end)
    formatted_dates = [date.strftime('%d/%m/%Y') for date in date_range]
    for i in formatted_dates:
        room = (int(str(no)[2:])-1)
        date_roomnumber[i][type_room][room] = 0
    return 
# room_no_to_zero('15/02/2024','18/02/2024','Standard Non-AC',403)

def customer_detail(room_type_l):
    # i create the list when we will store the data 
    global booking_detail
    global date_roomnumber,date_acc_room
    try:
        name_list = []
        age_list = []
        mob_no = []
        id_list = []
        date_in = []
        date_out = []
        booking_id = []
        amount = [] # its store payment transction id 
        room_type = []
        room_quantity = []
        booking_date = []
        status  = []
        room_number = [] 
        cust_city = []
        price_room = [] # its store price amount
        # generate the current date
        curr_date = datetime.datetime.now()
        date = curr_date.strftime('%d/%m/%Y')
    
        end_date = curr_date + datetime.timedelta(days=15)
        while 1:
                check_date = input(f"Enter Check IN Date({date} - {end_date.strftime('%d/%m/%Y')}): ")
                if curr_date <= datetime.datetime.strptime(check_date, '%d/%m/%Y') <= end_date:
                    while 1:
                        check_out = input("Enter Check Out Date: ")
                        if datetime.datetime.strptime(check_out, '%d/%m/%Y') > datetime.datetime.strptime(check_date, '%d/%m/%Y') :
                            available_room = room_check(check_date,check_out,room_type_l)
                            print(f"Available Room: {available_room} for {check_date}: ")
                            #customer Detail
                            name= input("Enter Your Name (0 for Exit): ").title()
                            if name == '0': return
                            else:     
                                while 1:
                                    age = (input("Enter Your Age (Book Only Adult Person ): "))
                                    if age.isnumeric() and len(age) == 2 and age >=  '18':
                                        while 1:
                                            mob = input("Enter Your Mobile Number: ")
                                            if mob.isnumeric() and len(mob) == 10:
                                                city = input('Enter Your City: ').upper()
                                                id = input("Enter Your ID NO: ").upper()
                                                # price calcultion
                                                price = room_price(room_type_l,check_date,check_out) 
                                                booking_i = book_id()
                                                con = input('GO TO PYMENT Y/N: ').upper()
                                                if con == 'N':return
                                                elif con == 'Y':
                                                     #room booking more detail
                                                    data = room_num_date(check_date,check_out,room_type_l)
                                                    d_room_no = [i for i in data if i!= 0]
                                                    room_num = rp.choice(d_room_no) # append the date left lists
                                                    room_no_to_zero(check_date,check_out,room_type_l,room_num)  # this is replace the actual value to zeros
                                                    room_no_subtract(check_date,check_out,room_type_l) # this ('-') values asign in every booked room 
                                                    age_list.append(age)
                                                    name_list.append(name)
                                                    mob_no.append(mob)
                                                    cust_city.append(city)
                                                    id_list.append(id)
                                                    date_in.append(check_date)
                                                    date_out.append(check_out) 
                                                    booking_date.append(date)
                                                    status.append('Booked')
                                                    room_quantity.append(1)
                                                    room_type.append(room_type_l)
                                                    room_number.append(room_num)
                                                    booking_id.append(booking_i)
                                                    pur = 'Room Booking'
                                                    transction = payment(booking_i,price,pur)
                                                    price_room.append(price)
                                                    amount.append(transction)
                                                    new_df = pd.DataFrame({
                                                        'Booking ID': booking_id,
                                                        'Name': name_list,
                                                        'Age': age_list,
                                                        'Check In': date_in,
                                                        'Check_Out': date_out,
                                                        'ID': id_list,
                                                        'Book Date': booking_date,
                                                        'Room NO': room_number,
                                                        'Room Type': room_type,
                                                        'NO. of Room': room_quantity,
                                                        'Transction ID': amount,
                                                        'Mobile': mob_no,
                                                        'City': cust_city,
                                                        'Status': status,
                                                        'Amount':price_room
                                                    })
                                                    
                                                    booking_detail = pd.concat([booking_detail, new_df], ignore_index=True)
                                                    print('****************Booking Successfully****************')
                                                    return
                                            else:print('Enter Correct Mobile Number')
                                    else: print('Enter Correct  Age')
                        else:print('Please Correct date') 
                else:print('Please Enter data in range')
    except Exception as e:
        print(f"An error occurred: {e}")
        print("Please retry.")

    return    
# customer_detail("3-Bed AC")

# this funcation is given the room number according to date ## cancel
def replace_cancel_roomnum(date,end,room_type_l,room_n):
    global date_roomnumber
    date_range = pd.date_range(start=date, end=end)
    formatted_dates = [date.strftime('%d/%m/%Y') for date in date_range]
    room = (int(str(room_n)[2:])-1)
    if date in date_roomnumber.keys():
        for date in formatted_dates: 
                date_roomnumber[date][room_type_l][room] = room_n
        return

        
# [401,402,403,404]

# this function is add room according to date  no of Room ### cancel
def can_add_room_data(date,end,room_type_l):
    global date_acc_room
    date_range = pd.date_range(start=date, end=end)
    formatted_dates = [date.strftime('%d/%m/%Y') for date in date_range]
    if date in date_acc_room.keys():
        for date in formatted_dates: 
            date_acc_room[date][room_type_l] +=1
        return
# 10

def cancel_booking():
        global booking_detail
        print(booking_detail)
        print('Are you Cancel Booking ')
        try:
            while 1:
                    booking_id_to_find = (input("Enter booking Id or press 0 foe Exit: ")).upper()
                    if booking_id_to_find == '0':
                        return 
                    elif 'Booking ID' in booking_detail.columns:
                        data = booking_detail.loc[(booking_detail['Booking ID'] == booking_id_to_find)]
                        if  not data.empty:
                            check_in = data['Check In'].values[0] 
                            check_out = data['Check_Out'].values[0]
                            room_no = data['Room NO'].values[0]
                            room_type = data['Room Type'].values[0]
                            status = 'Cancelled'
                            can_add_room_data(check_in,check_out,room_type)
                            replace_cancel_roomnum(check_in,check_out,room_type,room_no)
                            booking_detail.loc[(booking_detail['Booking ID'] == booking_id_to_find),'Status'] = 'Cancelled'
                            print('Cancelled Successfull')
                            return
        except Exception as e:
            print(f"An error occurred: {e}")
            print("Please retry.")

        return    
# cancel_booking()

# this function is add room according to date  no of Room (only for check availble)
def room_Avaible_for_date(date,room_type_l):
    global room_no_of,date_acc_room
    if date in date_acc_room.keys(): 
            return date_acc_room[date][room_type_l]
        
    else: 
        if date not in date_acc_room:
                date_acc_room[date] = room_no_of.copy()
                return date_acc_room[date][room_type_l]
# room_check('15/02/2024','Standard Non-AC') #10

def restaurant():
    global food_df, booking_detail
    try:
        tran_id = []
        id = []
        amount = []
        date = []
        d_ate = datetime.datetime.now().strftime('%d/%m/%Y')
        while 1:
            booking_id_to_find = input('Enter Customer id or 0 for exits: ')
            if not booking_detail.loc[booking_detail['Booking ID'] == booking_id_to_find].empty:   
                print("-------------------------------------------------------------------------")
                print("                           Hotel AnCasa")
                print("-------------------------------------------------------------------------")
                print("                            Menu Card")
                print("-------------------------------------------------------------------------")
                print("\n BEVERAGES                              26 Dal Fry................ 140.00")
                print("----------------------------------      27 Dal Makhani............ 150.00")
                print(" 1  Regular Tea............. 20.00      28 Dal Tadka.............. 150.00")
                print(" 2  Masala Tea.............. 25.00")
                print(" 3  Coffee.................. 25.00      ROTI")
                print(" 4  Cold Drink.............. 25.00     ----------------------------------")
                print(" 5  Bread Butter............ 30.00      29 Plain Roti.............. 15.00")
                print(" 6  Bread Jam............... 30.00      30 Butter Roti............. 15.00")
                print(" 7  Veg. Sandwich........... 50.00      31 Tandoori Roti........... 20.00")
                print(" 8  Veg. Toast Sandwich..... 50.00      32 Butter Naan............. 20.00")
                print(" 9  Cheese Toast Sandwich... 70.00")
                print(" 10 Grilled Sandwich........ 70.00      RICE") 
                print("                                       ----------------------------------")
                print(" SOUPS                                  33 Plain Rice.............. 90.00")
                print("----------------------------------      34 Jeera Rice.............. 90.00")
                print(" 11 Tomato Soup............ 110.00      35 Veg Pulao.............. 110.00")
                print(" 12 Hot & Sour............. 110.00      36 Peas Pulao............. 110.00")
                print(" 13 Veg. Noodle Soup....... 110.00")
                print(" 14 Sweet Corn............. 110.00      SOUTH INDIAN")
                print(" 15 Veg. Munchow........... 110.00     ----------------------------------")
                print("                                        37 Plain Dosa............. 100.00")
                print(" MAIN COURSE                            38 Onion Dosa............. 110.00")
                print("----------------------------------      39 Masala Dosa............ 130.00")
                print(" 16 Shahi Paneer........... 110.00      40 Paneer Dosa............ 130.00")
                print(" 17 Kadai Paneer........... 110.00      41 Rice Idli.............. 130.00")
                print(" 18 Handi Paneer........... 120.00      42 Sambhar Vada........... 140.00")
                print(" 19 Palak Paneer........... 120.00")
                print(" 20 Chilli Paneer.......... 140.00      ICE CREAM")
                print(" 21 Matar Mushroom......... 140.00     ----------------------------------")
                print(" 22 Mix Veg................ 140.00      43 Vanilla................. 60.00")
                print(" 23 Jeera Aloo............. 140.00      44 Strawberry.............. 60.00")
                print(" 24 Malai Kofta............ 140.00      45 Pineapple............... 60.00")
                print(" 25 Aloo Matar............. 140.00      46 Butter Scotch........... 60.00")
                print()
                r = 0
                while 1: 
                    ch=int(input(" Enter number in front of dish name or 0 for Exits:  "))
                     
                    # if-elif-conditions to assign item
                    # prices listed in menu card
                    if ch==1 or ch==31 or ch==32:
                        rs=20
                        r=r+rs
                    elif ch<=4 and ch>=2:
                        rs=25
                        r=r+rs
                    elif ch<=6 and ch>=5:
                        rs=30
                        r=r+rs
                    elif ch<=8 and ch>=7:
                        rs=50
                        r=r+rs
                    elif ch<=10 and ch>=9:
                        rs=70
                        r=r+rs
                    elif (ch<=17 and ch>=11) or ch==35 or ch==36 or ch==38:
                        rs=110
                        r=r+rs
                    elif ch<=19 and ch>=18:
                        rs=120
                        r=r+rs
                    elif (ch<=26 and ch>=20) or ch==42:
                        rs=140
                        r=r+rs
                    elif ch<=28 and ch>=27:
                        rs=150
                        r=r+rs
                    elif ch<=30 and ch>=29:
                        rs=15
                        r=r+rs
                    elif ch==33 or ch==34:
                        rs=90
                        r=r+rs
                    elif ch==37:
                        rs=100
                        r=r+rs
                    elif ch<=41 and ch>=39:
                        rs=130
                        r=r+rs
                    elif ch<=46 and ch>=43:
                        rs=60
                        r=r+rs
                    elif ch == 0:
                        
                        con = input('GO TO PYMENT Y/N: ').upper()
                        if con == 'N':return
                        elif con == 'Y':
                            id.append(booking_id_to_find)
                            amount.append(r)
                            date.append(d_ate)
                            pur = 'Food Order'
                            transction = payment(booking_id_to_find,r,pur)
                            tran_id.append(transction)
                            df_res = pd.DataFrame({'Transction ID':tran_id,'Customer ID': id, 'Date': date,'Total_Amount':amount})
                            food_df = pd.concat([food_df, df_res], ignore_index=True)
                            print(f'''            ---------------------------------------------------------
                            Hotel AnCasa
            ---------------------------------------------------------
                                        BILL 
            ---------------------------------------------------------
            Customer ID {booking_id_to_find}     Date {d_ate}
            ---------------------------------------------------------
                                    Total Amount                                     
                                        {r} 
            ---------------------------------------------------------
            Thank you For Order.Please Visit Again
            ---------------------------------------------------------''')
                            return
                    else:
                        print("Wrong Choice..!!")
            elif booking_id_to_find == '0':return 
            else: 
                print('Wrong Customer ID')  
    except Exception as e:
        print(f"An error occurred: {e}")
        print("Please retry.")

    return 
# restaurant()


def Room_Avialbilty(type_room):
    global date_acc_room
    try:
        curr_date = datetime.datetime.now()
        date = curr_date.strftime('%d/%m/%Y')
    
        end_date = curr_date + datetime.timedelta(days=15)
        while 1:
                print('0 for Exits')
                check_date = input(f"Enter Check IN Date({date} - {end_date.strftime('%d/%m/%Y')}): ")
                if curr_date <= datetime.datetime.strptime(check_date, '%d/%m/%Y') <= end_date:
                    room = room_Avaible_for_date(check_date,type_room)
                    print(f"Available Room: {room} for {check_date}: ") 
                    return
                else: print('Invalid Input')
    except Exception as e:
        print(f"An error occurred: {e}")
        print("Please retry.")

    return
# Room_Avialbilty('Standard Non-AC')

def booking_history():
        global booking_detail
        print(booking_detail)
        while 1:
                booking_id_to_find = (input("Enter booking Id or press 0 foe Exit: ")).upper()
                if booking_id_to_find == '0':
                    return 
                elif 'Booking ID' in booking_detail.columns:
                    data = booking_detail.loc[(booking_detail['Booking ID'] == booking_id_to_find)]
                    print(data) 

# ROOMS INFO 
def Rooms_Info():
    print("         ------ HOTEL ROOMS INFO ------")
    print("")
    print("STANDARD NON-AC")
    print("---------------------------------------------------------------")
    print("Room amenities include: 1 Double Bed, Television, Telephone,")
    print("Double-Door Cupboard, 1 Coffee table with 2 sofa, Balcony and")
    print("an attached washroom with hot/cold water.\n")
    print("STANDARD NON-AC")
    print("---------------------------------------------------------------")
    print("Room amenities include: 1 Double Bed, Television, Telephone,")
    print("Double-Door Cupboard, 1 Coffee table with 2 sofa, Balcony and")
    print("an attached washroom with hot/cold water + Window/Split AC.\n")
    print("3-Bed NON-AC")
    print("---------------------------------------------------------------")
    print("Room amenities include: 1 Double Bed + 1 Single Bed, Television,")
    print("Telephone, a Triple-Door Cupboard, 1 Coffee table with 2 sofa, 1")
    print("Side table, Balcony with an Accent table with 2 Chair and an")
    print("attached washroom with hot/cold water.\n")
    print("3-Bed AC")
    print("---------------------------------------------------------------")
    print("Room amenities include: 1 Double Bed + 1 Single Bed, Television,")
    print("Telephone, a Triple-Door Cupboard, 1 Coffee table with 2 sofa, ")
    print("1 Side table, Balcony with an Accent table with 2 Chair and an")
    print("attached washroom with hot/cold water + Window/Split AC.\n\n")
    print()
    while 1:
        n=int(input("0-BACK\n ->"))
        if n==0:
            return
        else:
           print('Wrong Input') 
# Rooms_Info()

def check_room_Avialbilty():
    while 1:
        print("----SELECT ROOM TYPE FOR CHECK ROOM AVAILABILITY----")
        print(" 1. Standard Non-AC")
        print(" 2. Standard AC")
        print(" 3. 3-Bed Non-AC")
        print(" 4. 3-Bed AC")
        print(" 0. for Exit")
        mode = int(input("Select Any one Opition: "))
        if mode == 1:
            room_type  = 'Standard Non-AC'
            Room_Avialbilty(room_type)
        elif mode == 2:
            room_type  = "Standard AC"
            Room_Avialbilty(room_type)
        elif mode == 3:
            room_type  = "3-Bed Non-AC"
            Room_Avialbilty(room_type)
        elif mode == 4:
            room_type  = "3-Bed Non-AC"
            Room_Avialbilty(room_type)
        elif mode == 0:
            return


def room_booking():
    while 1:
        print("----SELECT ROOM TYPE----")
        print(" 1. Standard Non-AC")
        print(" 2. Standard AC")
        print(" 3. 3-Bed Non-AC")
        print(" 4. 3-Bed AC")
        print(" 0. for Exit")
        mode = int(input("Select Any one Opition: "))
        if mode == 1:
            room_type  = 'Standard Non-AC'
            customer_detail(room_type)
        elif mode == 2:
            room_type  = "Standard AC"
            customer_detail(room_type)
        elif mode == 3:
            room_type  = "3-Bed Non-AC"
            customer_detail(room_type)
        elif mode == 4:
            room_type  = "3-Bed Non-AC"
            customer_detail(room_type)
        elif mode == 0:
            return 
# room_booking()

def main():
    while 1:
        print("""Press 1 For Booking 
press 2 For Room Avialbilty
press 3 For Room Information
press 4 For Room Service
press 5 For Booking History
press 6 For Cancel booking
press 0 For Exit
    """)
        Mode = int(input("Select Any One Option: "))
        if Mode ==1:
            room_booking() # complete
        elif Mode == 2:
            check_room_Avialbilty() # complete
        elif Mode == 3:
            Rooms_Info() # complete
        elif Mode == 4:
            restaurant() # complete
        elif Mode == 5:
            booking_history() # complete
        elif Mode == 6:
            cancel_booking()# complete
        elif Mode == 0:
            return
main()        

booking_detail


food_df


df_payment



```
